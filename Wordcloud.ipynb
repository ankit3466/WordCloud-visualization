```python
import wikipedia
from wordcloud import WordCloud , STOPWORDS

from PIL import Image
import numpy as np
```

```python
#getting data
text = "Python programing language is one of the most famous language used for Artificial Intelligence."
```

```python
# wordcloud object

wc = WordCloud(max_words=10,
              stopwords=None)
```

```python
# generate visual

x = wc.generate(text)
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(x)
plt.axis("off")
plt.show()
```
