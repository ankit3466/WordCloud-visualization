```python
import wikipedia
from wordcloud import WordCloud , STOPWORDS

from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
# getting data

def get_data(topic):
    
    title = wikipedia.search(topic)[0]
    page = wikipedia.page(title)
    return page.content
```

```python
text = get_data("Python language")
```

```python
# preview of data
text[:100] , len(text)
```

```python
# stopword

stopword = set(STOPWORDS)
len(stopword)
```

```python
# take the mask

image = Image.open('C://Users/user/OneDrive/Desktop/washington.jpg')
```

```python
mask = np.array(image)
```

```python
plt.imshow(image)
```

```python
# wordcloud object

wc = WordCloud(max_words=100,
              stopwords=stopword,
              background_color='black',
              mask=mask)
```

```python
wc.generate(text)
```

```python
# saving the image

wc.to_file('word-cloud.png')
```

```python
plt.imshow(wc)
plt.axis("off")
plt.show()
```
