```python
import wikipedia
from wordcloud import WordCloud , STOPWORDS

from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
# getting data

def get_data(topic):
    
    title = wikipedia.search(topic)[0]
    page = wikipedia.page(title)
    return page.content
```

```python
text = get_data("Python language")
```

```python
# preview of data
text[:100] , len(text)
```

```python
# stopword

stopword = set(STOPWORDS)
len(stopword)
```

```python
# wordcloud object

wc = WordCloud(max_words=200,
              stopwords=stopword,
              background_color='black',
              )
```

```python
# generate visual

x = wc.generate(text)
```

```python
plt.imshow(x)
plt.axis("off")
plt.show()
```
