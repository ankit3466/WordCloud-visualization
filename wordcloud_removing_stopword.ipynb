```python
import wikipedia
from wordcloud import WordCloud , STOPWORDS

from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
#getting data
text = "Python programing language is one of the most famous language used for Artificial Intelligence is is is is"
```

```python
# stopwords
stopword = set(STOPWORDS)
```

```python
# wordcloud object

wc = WordCloud(max_words=10,
              stopwords=stopword,
              background_color='white',
              relative_scaling=1.0)
```

```python
# generate visual

x = wc.generate(text)
```

```python
plt.imshow(x)
plt.axis("off")
plt.show()
```
